## PROMPTING

Muchos han usado Chat GPT u otro modelo, cierto?, pero lo más probable es que las interacciones hayan sido usando la interfaz visual, el Chat. Si bien las mismas técnicas se pueden usar en otros modelos (Gemini, Mistral, etc) hay variaciones en los endpoints y métodos. En este ejercicio exploraremos una interacción diferente, conectándonos a la api de OPenAI usando Pyrthon.  

**Parámetros** Para usar la api hay que considerar algunos parámetros para trabajar. Algunos de los más importantes son:
- engine: es el nombre del modelo entrenado a usar, por ej GPT-4o, GPT-4omini, GPT-Turbo, etc.
- prompt: es la pregunta, consulta o solicitud a realizar generalmente en una ventana de chat.   
- n: es el número de respuestas solicitadas, 1 por ejemplo.   
- temperatura: modula consistencia y novedad del resultado. Valores menores dan resultados más uniformes y previsibles (deterministas). Valores más altos hacia mayor diversidad y creatividad en respuestas (pero podría alucinar). Calibrar  la temperatura en función del equilibrio deseado entre coherencia e innovación.
- max_tokens: máximo de tokens en la respuesta, usar de 2xn, como 64, 128, 256, 512, 1024, 2048, con un máximo de 4096.   

**Roles** En la interacción podemos mejorar las respuestas dando características a los roles disponibles, existen tres roles: system, user, assistant. Al menos en la API de OpenAI, hay roles para guiar las respuestas del modelo, esto ayuda a definir el contexto orientando mejor la conversación. El uso estratégico de roles puede mejorar significativamente el resultado del modelo.

Los más comunes:   

- "system": proporciona instrucciones generales, entrega instrucciones iniciales para guiar el comportamiento, tono y actitud del sistema para generar la respuesta y calzar con las necesidades específicas. Se evalúan una vez al inicio de la conversación.   

- "user": presenta consultas o indicaciones, es quien realiza las preguntas, instrucciones, o solicitudes en forma de prompts en la ventana del chat. Es quien conduce la conversación, o sea usted.   

- "assistant": es la respuesta del modelo o IA a los mensajes del USER, en forma de sugerencias explicaciones u otras. Responde a los parámetros del SYSTEM y del USER.

## B- Prompt inicial: Consulta simple.     

Estos modelos funcionan mediante una interfaz de conversación estructurada, y es de esta manera que se ingresan las consultas al sistema. Algunas recomendaciones:   
–Dar un contexto claro al rol: Comenzar con mensaje del sistema para definir el contexto o comportamiento del modelo, es una guía para las interacciones.   
–Indicaciones explícitas: ser claro y conciso en el rol del usuario ayuda al modelo a dar respuestas más precisas.    
–Retroalimentación: ante respuesta no satisfactoria, usar el rol de usuario para dar comentarios y refinar la consulta.   
–Conversación iterativa: Pensar como un diálogo, el modelo puede referenciar mensajes anteriores, conservando del contexto.   
Este motor de conversación funciona mediante una serie de mensajes con la siguiente estructura:   
```
{  
"role": "user",   
"content": "message"  
}
```

````python
## Primer prompt para probar la conexión.
## Poco a poco veremos la estructura del promt.
## con el rol de system, podemos darle personalidad a la respuesta

## mensaje para prompt, descomentar la variable a probar
quien_seras = "Eres un ganster muy malo, contestame muy mala onda"  # esto dara un contexto
#quien_seras = "Eres un niño de 3 años muy bueno e inocente, contestame muy buena onda"

un_prompt  = "Hola, mi nombre es Juanito, ¿quien eres tu?"

## establecer parametros...
primer_mensaje = [
  { "role": "system", "content": quien_seras },
  { "role": "user", "content": un_prompt }
]

## le enviamos el primer mensaje...
## completions "arma el paquete" con mensaje y parametros a enviar al modelo
completions0 = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    temperature = 0.9, # ente 0 y 2
    messages = primer_mensaje
)

## Al recibir la respuesta, mostramos la primera
respuesta0 = completions0.choices[0].message.content
respuesta0
````

````python
## Mostrar respuesta en bruto, observar que tiene varios elementos
completions0.choices[0]
````

## D- Resúmenes:
Un uso frecuentes es solicitar un resumen de un texto según extensión.
Usaremos el texto de la Araucana de Alonso de Ercilla. La obtendremos de Wikimedia, para su práctica puede copiar y pegar, usar cualquier canto. En este ejemplo usaremos un fragmento del Canto I

Además para este ejemplo ya usaremos una función para "empaquetar" las respuestas. La función se llamará: obtener_respuesta()


````python
## Función "obtener_repsuesta" permite "empaquetar" la consulta para no repetir tanto codigo
## Recordar correr esta celda para tenerla en memoria.

## la función tiene por parámetros: el prompt, un modelo y la temperatura.
## devuelve o return el contenido del mensaje de respuesta.

def obtener_respuesta(el_prompt, el_modelo="gpt-3.5-turbo", la_temperatura=0.8):
    el_mensaje2 = [
        { "role": "user",
        "content": el_prompt }
    ]
    el_completions2 = client.chat.completions.create(
        model = el_modelo,
        temperature = la_temperatura,
        messages = el_mensaje2,
    )
    return el_completions2.choices[0].message.content

## Una prueba de la función con una pregunta simple.
## Descomentar la pregunta que le interese.

respuesta2 = obtener_respuesta("Hola, me llamo Don Ramón. ¿Quién eres tu?")
#respuesta1 = obtener_respuesta("Hola, me llamo Don Ramón. ¿Dónde se inventaron las papas fritas?")
#respuesta2 = obtener_respuesta("Hola, me llamo Don Ramón. ¿Me puedes dar los números ganadores de la loteria de mañana?")

## probar respuesta de la funcion
print( respuesta2 )
````

## F- Extraer entidades con Cadena de pensamiento.   
Un muy buen uso que se puede dar a los prompts es extraer las entidades (personas, lugares, etc) y relaciones (pertenece A, pago A, viajo Hacia, dueño De, etc) presentes en un texto. Permite reconocer y extraer personas, objetos, fechas, lugares.
Se puede usar una estrategia conocida como Chain of Though (Cadena de pensamiento), que consiste en darle ciertas instrucciones básicas a seguir y obtener así un resultado deseado. Esto ayuda al sistema a entender las solicitudes paso a paso, logrando buscar en los embeddings del modelo los elementos más adecuados para la respuesta. En el fondo no piensa, busca texto en base a similitudes estructurales.   


````python
## Para extraer entidades usaremos un artículo de la Constitución, obtenido desde c80.cl
## copiar y pegar en el buscador https://c80.cl/wp-json/constitucion1980/v1/articulo/24

articulo_const = """Artículo 24.
El gobierno y la administración del Estado corresponden al Presidente de la República, quien es el Jefe del Estado.
Su autoridad se extiende a todo cuanto tiene por objeto la conservación del orden público en el interior y laseguridad externa de la República, de acuerdo con la Constitución y las leyes."
El 21 de mayo de cada año, el Presidente de la República dará cuenta al país del estado administrativo y político de la Nación ante el Congreso Pleno.
"""
````
