## 0- ¿Qué son las redes? (0 punto):

No se trata ni de redes sociales como facebook, ni de redes de pesca, ni redes de computadores. Es una ciencia que aparece el siglo 20 que describe y analiza colecciones de nodos y vínculos. LAs principales ciencias que las usan son la física, la sociología, y las matemáticas.

Podemos pensar ua red como un conjunto de Nodos y Vínculos:
    
**NODOS:** Son los elementos que tienen conexiones, pueden ser personas, cosas, palabras, etc. Pueden tener atributos como color de pelo, tamaño, u otros.  

**Vínculos:** Son las conexiones o tipos de relación entre los nodos, pueden ser por parentezco, amistad, etc. Estas relaciones pueden tener dirección o no. Por ejemplo no dirigido es si Juan y Pedro son compañeros de clase pues estarían al mismo nivel jerárquico. Pero si Juan prestó dinero a Pedro, esa relación tiene dirección, pues hay un iontercambio donde algo fluye de un nodo hacia otro.

**TIPOS DE GRAFOS:**  

Las conexiones o interacciones en las redes pueden tener o no una dirección, por lo que pueden ser dirigidas o no dirigidas.    
**No Dirigidas, o Simétrica:** Relación que no implica dirección o jerarquía, como una relación de amistad, o si son compañeros de curso.

**Dirigidas, o Asimétrica:** Implica dirección de la relación, la cual puede estar dada por jerarquía o dirección de la acción. Por ejemplo a quien se le presta dinero, a quien pregunta, a qué ciudad pertenece alguien, a quien se suministra algo, etc.   

**Componentes de un grafo** También un grafo puede estar completamente conectado (todos sus elementos relacionados) por lo que tiene un solo componente. A su vez, puede tener algunos elementos desconectados, por lo cual está compuesto por dos o más sub redes o componentes, desconectados entre ellos.   

**METRICAS:**

Para comprender la estructura de las redes se usan diferentes métricas. Las **métricas de centralidad** de los nodos es importante, pues permite ver cuáles son los nodos centrales de la red. Hay 3 tipos de centralidad:

**Centralidad de grado (degree centrality):** Es el nodo que tiene más conexiones a otros nodos. Indica el número de relaciones o con cuántos otros nodos se conecta. Permite ver qué tan importante es un nodo. Por ejemplo, con cuántas personas se conecta un individuo.  

**Centralidad de cercanía (closeness centrality):** Indica qué tan cercano está un nodo de otros nodos. Ayuda a reconocer el camimo más corto o más largo entre los nodos.   

**Centralidad de intermediación (betweenness centrality):** Indica qué tantas conexiones se permite en el grafo. Ayuda a ver cuales son los nodos que sirven como puentes entre sub grafos. Por ejemplo, una persona que conecta dos grupos diferentes de personas es un intermediador.

Otras medidas son:  
**Distancia:** Número de pasos de separación entre nodos. Entre dos nodos, cuántos nodos intermedios hay.   

**Distancia o ruta geodésica:** Señala la distancia usando el camino más corto entre nodos. Son el número de enlaces en la ruta más corta que los conecta.  

**Diámetro Máximo:** Es la distancia máxima entre cualquier par de nodos en el grafo. De todas las distancias, cual es la máxima.

## B- Redes: Crear nodos y enlaces (0 puntos):  

Si no se tiene un archivo de grafos, se puede construir un grafo. Para esto se crean las entidades y las relaciones. Hay muchas maneras, pero veremos una de las más simples.

A continuación un ejemplo de como crear una red de amigos. De cada persona se vincula con quien tiene una amistad (de a pares), y se guardan en una variable cada par. Luego, se convierte en Dataframe de Pandas. Finalmente se convierte el dataframe en una red de NetworkX.  


```python
## - - - - - - - - - - - -
## B1- Redes: Estructura de datos.
## - - - - - - - - - - - -

## Creación de relaciones: cada persona se coloca con quien tiene una amistad (de a pares)
## Se guardan en una variable cada par.
## Es una red no direccionada, ["Alicia","Beatriz"] es lo mismo que ["Beatriz","Alicia"]
## Por eso cuidar de no repetir las relaciones.

p01 = [ "Alicia","Beatriz" ]
p02 = [ "Alicia","Carlos" ]
p03 = [ "Alicia","Diego" ]
p04 = [ "Alicia","Enrique" ]
p05 = [ "Alicia","Ignacia" ]
p06 = [ "Alicia","Sergio" ]
p07 = [ "Beatriz","Sergio" ]
p08 = [ "Beatriz","Tania" ]
p09 = [ "Beatriz","Ursula" ]
p10 = [ "Carlos","Ignacia" ]
p11 = [ "Carlos","Victor" ]
p12 = [ "Diego","Karen" ]
p13 = [ "Diego","Leo" ]
p14 = [ "Diego","Jose" ]
p15 = [ "Diego","Enrique" ]
p16 = [ "Diego","Felipe" ]
p17 = [ "Enrique","Beatriz" ]
p18 = [ "Enrique","Ximena" ]


# Nombres de variables en listas para hacerlo más facil.
personas = [ p01, p02, p03, p04, p05, p06, p07, p08, p09, p10,
             p11, p12, p13, p14, p15, p16, p17, p18  ]
personas_cols = [ "p01", "p02", "p03", "p04", "p05", "p06", "p07", "p08", "p09", "p10",
                  "p11", "p12", "p13", "p14", "p15", "p16", "p17", "p18" ]

# Se convierten las relaciones de las listas en Dataframe de Pandas
# Más adelante se convierte el DataFrame en una red de NetworkX.
dfred = pd.DataFrame( personas )
dfred.index = personas_cols

# Mostrar el DataFrame de Pandas
dfred
```

```python
# Finalmente se convierte el DataFrame en una red de NetworkC.

# Al grafo de los amigos le llamaremos GA
GA = nx.from_pandas_edgelist( dfred, 0, 1 ) # en GA guardamos los datos creados
```
