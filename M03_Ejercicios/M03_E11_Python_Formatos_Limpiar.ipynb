## C - Limpieza de datos

Al inspeccionar los datos de ejemplo (ciej_utiles_sucio.csv), detectamos algunos problemas: muchos valores vacíos o NaN, otros mal escritos, incosistencia de nombres y fechas, etc. Detectar los problemas permite definir las operaciones a realizar para la limpieza.   

En la etapa de limpieza por lo general se realizan una serie de operaciones (dependiendo de los datos, del problema o del escenario, pueden requerirse otras operaciones):  
* formatear los datos unificando valores con puntos o comas.  
* detectar filas con más NaN (o Not a Number).  
* informacion incompleta que en ciertos casos se completa o elimina. 
* se colocan textos en baja.  

Se recomienda que el resultado de la limpieza se guarde en un nuevo archivo para ser usado en etapas siguientes. Realizaremos algunas operaciones. 

**Si necesitan botar FILAS vacias o que no sirven:**    
- Indicar que filas: [desde:hasta]  
df = df.drop( df.index[11:955] )    
df # siempre mostrar los resultados para verificar el resultado

**Si necesitan botar COLUMNAS vacias o que no sirven:**    
Indicar una lista de que columnas botar: ["esta", "y esta"]   
df = df.drop(['Unnamed: 6', 'Unnamed: 7'], axis=1)   
df # siempre mostrar los resultados para verificar el resultado   


```python
# C.1- Con esta línea puedes botar una fila que no sirva o tenga muchos valores NAN
# en el ejemplo botaremos la fila 5 pues tiene muchos elementos nulos. 
# ojo: no correrla muchas veces, sino botar más filas. 
# De ser asi, repetir los pasos anteriores nuevamente, desde la carga de datos.

# OJO: asegurarse de botar la fila con mas NaN. En el ejempñlo la fila 5 
df = df.drop( df.index[ 5 ] )  
df
```

```python
# C.2- Fijarse en el index
# Si al botar una o más filas, el index queda con números desordenados o falta de alguno 
# Se puede resetear el index con estas líneas

df.reset_index(inplace = True)
df
```

```python
# C.3- Para continuar la limpieza, vamos a copiar df en otro dataframe llamado dflimpiando.
# Esto para ir separando los procesos.  
# nos crea otra columna de index que podemos borrar si lo estimamos. 

df = df.drop(['index'], axis=1)   
dflimpiando = df.copy()

# mostrar dataframe 
dflimpiando
```

```python
# C.4- En la última columna del ejemplo hay una o más celdas que sin un valor el precio. 
# Podemos agregar un valor de la misma columna. agrgaremos 135
# asegurarse que sea la fila y columna correctas

dflimpiando.at[7, 'costounidad'] = 135

# mostrar dataframe 
dflimpiando
```

```python
# C.5- Ahora veremos como unificar valores para que sean consistentes.  
# Se identifican los valores únicos en una columna: nombre e item
# para ver los valores mal escritos o inconsistentes y así poder unificar. 

dflimpiando['item'].unique()  
```

```python
# C.6- Con esa información, vemos que en dicha columna hay inconsistencias: 
# 'LAPI', 'LAPICS', 'lapiz', deben ser 'lápiz'
# 'LIBRO', 'libro', deben ser 'libro'
# Por ahora reemplazaremos "a mano" valores inconsistentes por los correctos 
# Para esto usatmremos un diccionario con palabras 'incorrecta' : 'correcta'

dflimpiando = dflimpiando.replace( {'item':
                                    { 'LAPI':'lápiz',
                                      'LAPICS':'lápiz',
                                      'lapiz':'lápiz',
                                      'LIBRO':'libro'}
                                   } )
 
# mostrar dataframe 
dflimpiando
```

```python
# C.7- Generalmente se unifican los nombres, y se ponen en bajas
# Seleccionar la columna a unificar y poner valores en bajas con las funciones .str.lower() 

dflimpiando['nombre'] = dflimpiando['nombre'].str.lower() 

# mostrar dataframe 
dflimpiando
```

```python
# C.8- Las fechas están en diferentes formatos, se debe unificar. 

dflimpiando['fecha'] = pd.to_datetime(dflimpiando.fecha)

# mostrar dataframe 
dflimpiando
```

## D - Datos derivados

Los datos ahora estan en mejor forma, pero podemos crear datos nuevos, derivándolos de datos existentes. En el ejemñplo, podemos hacer una columna derivada con el total de los valores. 


```python
# D.1- Crear una nueva columna en base a otras (de la cantidad y el precio unitario) 
# total = unitario x cantidad 
# dflimpiando['total'] crea la columna nueva con el resultado de la operación entre otras dos columnas:  

dflimpiando['total'] = dflimpiando['cantidad'] * dflimpiando['costounidad']

# mostrar dataframe 
dflimpiando
```

```python
# D.2- Ahaora tenemos los datos algo más ordenado y prácticamente limpios. 
# Mostramos el dflimpiando para ver como va quedando.

dflimpiando 
```
