## D - Visualizar scatterplot 1 y 4 conjuntos

Un gráfica de dispersión (burbujas, bolas, etc) es un tipo de diagrama con coordenadas cartesianas, y permite mostrar dos variables de un conjunto de datos. En este scaterpot graficaremos los 4 conjuntos o set de datos, I, II, III y IV.


```python
## D1- Primero como ejemplo haremos un scatterplot de un solo conjunto de datos.

## Datos seleccionados del conjunto I
x1 = df['I_x']
y1 = df['I_y']

## Tamaño y proporción del gráfico.
fig = plt.figure(figsize=(6, 4))

plt.title("Cuarteto de Anscombe x1 e x1 x de serie I") # título.
plt.xlabel("El Eje X") # texto eje X.
plt.ylabel("El Eje Y") # texto eje Y.

plt.scatter(x=x1, y=y1, color='red') # gráfico de puntos.
plt.show()
```

```python
## D2- En este scaterpot graficaremos los 4 conjuntos (sets) de datos, I, II, III y IV.
## Seleccionar los datos, colcarlos en el gráfico correspondiente, poner títulos, y seleccionar diferentes colores.

x1, y1 = df['I_x'], df['I_y']
x2, y2 = df['II_x'], df['II_y']
x3, y3 = df['III_x'], df['III_y']
x4, y4 = df['IV_x'], df[ 'IV_y' ]

## Tamaño y proporción del gráfico.
fig=plt.figure(figsize=(16, 10)) # tamaño del gráfico.

## grafica el primer conjunto. 
plt.subplot(221) # 2 columnas, 2 filas y posición 1.
plt.title('Serie I')
plt.scatter(x1, y1, s=80, color='red', marker='o' )

## grafica el segundo conjunto. 
plt.subplot(222) # 2 columnas, 2 filas y posición 2.
plt.title('Serie II')
plt.scatter(x2, y2, s=80, color='blue', marker='o')

## grafica el tercer conjunto. 
plt.subplot(223)
plt.title('Serie III') # 2 columnas, 2 filas y posición 3.
plt.scatter(x3, y3, s=80, color='orange', marker='o')

## grafica el cuarto conjunto. 
plt.subplot(224) # 2 columnas, 2 filas posición 4.
plt.title('Serie IV')
plt.scatter( x4, x4 , s=80, color='green', marker='o')

plt.show()
```

```python
## D3- Describir resumen estadístico.
## Atención a los graficos:
## Lo resúmenes estadísticos dicen que todos los datos son iguales
## Pero los gráficos muestran otra cosa...

df.describe() # fiajarse en ""mean", "std" de los x y los y, varios son iguales
```

### Comparación estadísticos (describe) y gráficos

Observe y compare el promedio (mean), la desviación estandar (std) de los 4 conjuntos. los x son iguales, los y son iguales.

Según estas estadísticas, se podria decir que los 4 se comportan igual. Pero los gráficos muestran distribuciones diferentes. La importancia de la visualización radica en que muestra patrones de comportamiento de los datos, lo que permite detectar como se distribuyen, como se agrupan, y si existen outliers (datos que se escapan de los grupos) 


## E - Visualizar scatterplot 4 conjuntos en un plot

En este scaterpot graficaremos dos conjuntos de datos, I y II.
como ejercicio puede incorporar las  series III y IV y adaptar colores y titulos.

```python
## E1- Scatterplot oi grafico de puntos (pueden tener otras formas tb)

## Datos - - - - - - - - - - - -
## guardamos datos en variables
g1 = (df['I_x'], df['I_y'])
g2 = (df['II_x'], df['II_y'])
g3 = (df['III_x'], df['III_y'])
g4 = (df['IV_x'], df['IV_y'])

gdatos   = (g1, g2, g3, g4 ) # guardamos esos grupos en la variable gdatos para acceder facilmente.
gnombres = ('El grupo 1', 'El grupo 2', 'El grupo 3', 'El grupo 4') # guardamos nombres de grupos
gcolores = ('red', 'blue', 'green', 'orange') # guardamos colores, puede cambiarlos: otors colores: https://matplotlib.org/stable/gallery/color/named_colors.html
gmarker  = ('o', 'v', 'd', 's'  ) # markers o formas de puntos, aca mas: https://matplotlib.org/stable/api/markers_api.html


## Crear gráficos - - - - - - - - - - - -
fig=plt.figure(figsize=(10, 6), facecolor='lightslategray') # tamaño del gráfico.
fig.patch.set_facecolor('green') # color del margen.
fig.patch.set_alpha(0.2) # transparencia del margen.

ax = fig.add_subplot(1, 1, 1) # agregar sub gráficos en el mismo espacio.
ax.patch.set_color('yellow') # color del gráfico.
ax.patch.set_alpha(0.9) # transparencia del gráfico.

plt.title("El Famoso Cuarteto de Anscombe") # titulo general.
plt.xlabel("Este es el Eje X") # texto eje X.
plt.ylabel("Este es el Eje Y") # texto eje Y.
plt.grid( color='grey', alpha=0.5) # grilla.

## se recorre la info de gdatos, gcolores, gnombres y se guardan en dat, col, gru.
for dat, col, gru, mar in zip(gdatos, gcolores, gnombres, gmarker):
    x, y = dat
    ax.scatter(x, y, alpha=0.6, c=col, edgecolors='none', s=80, label=gru, marker=mar) # puntos.
    plt.legend(gnombres) # leyendas.


## Mostrar grafico 
## SOLO COMPU LOCAL guarda el grafico en la ruta seleccionada.
#plt.savefig('a_cuarteto_scatter_series.jpg') # le piuede aparecer en la columna de al lado en los archivos

## DESCOMENTAR SOLO EN COLAB
#files.download("a_cuarteto_scatter_series.jpg")

## esta linea muestra el grafico.
plt.show()
```

```python
## F.1- Seleccionar articulo wikipedia

## Selecciona el idioma del artículo de WP.
wikipedia.set_lang("es")

## Ejemplo
## cual articulo a escoger: Visitar esta URL de ejemplo: https://es.wikipedia.org/wiki/La_Araucana
ejemplo_articulo_wp = 'La_Araucana'
articulo = wikipedia.page( ejemplo_articulo_wp )

## De las siguientes descomentar la que desee ver.

#print(articulo.original_title) # titulo del articulo
print( articulo.summary ) # sumario del articulo
print(articulo.content) # contenido textual del articulo
#print( articulo.categories ) # categorias del articulo
#print( articulo.images ) # lista url de imagenes del articulo
#print( articulo.url ) # url del articulo
#print(articulo.links[:]) # links que aparecen en el articulo
#print(articulo.html) # muestra html del articulo
```
