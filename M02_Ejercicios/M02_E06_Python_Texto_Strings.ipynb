## A - Strings y tipos de datos

Los datos pueden ser de distintos tipos:

- `int`: números enteros como 1, 55, -3
- `float`: números decimales como 3.4, 12.5
- `bool`: valores lógicos, True o False
- `str`: cadenas de texto como "hola", "diseño", "123"

Un número escrito con comillas es un string, no un número. Por ejemplo, `"5"` y `5` son distintos: el primero es texto, el segundo es un entero.

Para saber de qué tipo es un dato se usa `type()`.

```python
# Ejemplo A - tipos de datos
es_string   = "5"
es_entero   = 5
es_flotante = 5.2

print( type(es_string) )    # str
print( type(es_entero) )    # int
print( type(es_flotante) )  # float
```

## B - Indexación y slicing

Un string es una secuencia de caracteres. Cada caracter tiene una posición llamada index, partiendo desde 0.

Para acceder a un caracter: `texto[index]`. El index puede ser negativo: `-1` es el último caracter.

Para seleccionar un rango se usa slicing: `texto[inicio:fin]`. El caracter en `fin` no se incluye. Si se omite `inicio` o `fin`, se toma desde el principio o hasta el final.

```python
# Ejemplo B - indexación y slicing
txt = "diseño"

print( txt[0] )    # primer caracter: d
print( txt[-1] )   # último caracter: o
print( txt[0:3] )  # del index 0 al 2: dis
print( txt[3:] )   # desde index 3 hasta el final: eño
```

## C - Métodos de string

Un método es una función que se aplica a un dato con la notación de punto: `string.metodo()`.

Algunos métodos útiles:

- `.upper()` convierte a mayúsculas
- `.lower()` convierte a minúsculas
- `.strip()` elimina espacios al principio y al final
- `.replace(viejo, nuevo)` reemplaza un texto por otro
- `.find(texto)` devuelve el index donde comienza el texto buscado
- `.split(separador)` divide el string en una lista según un separador
- `.join(lista)` une los elementos de una lista en un string

```python
# Ejemplo C - métodos de string
frase = "Esta Frase, de Juan, es muy Buena     "

print( frase.upper() )                   # a mayúsculas
print( frase.lower() )                   # a minúsculas
print( frase.strip() )                   # sin espacios al final
print( frase.replace("Juan", "Bart") )   # reemplaza nombre
print( frase.find("Frase") )             # posición de "Frase": 5

partes = frase.split(",")               # divide por coma
print( partes )                          # lista de partes

nombres = ("Aldo", "Paula", "Gato")
print( "#".join(nombres) )              # Aldo#Paula#Gato
```

## D - Casting y f-strings

Casting es convertir un dato de un tipo a otro:

- `int(x)` convierte x a entero
- `float(x)` convierte x a flotante
- `str(x)` convierte x a string

Un f-string permite insertar variables directamente en un texto. Se escribe con `f"..."` y las variables van dentro de `{}`.

```python
# Ejemplo D - casting y f-strings
na   = int("5")  # string "5" se convierte en el entero 5
nb   = int("3")
suma = na + nb

print( f"Primero el {na}, luego el {nb}. Y estos suman {suma}" )
```
