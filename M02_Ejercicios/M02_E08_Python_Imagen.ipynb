## A- Imagen: Imagen es una matriz. 

Para empezar debemos considerar que una imagen es una matriz de valores. Veamos la diferencia entre vector y matriz: 

**Vector:** Es una lista de elementos, como las listas que hemos visto en python, se extiende en una dimensión, por ejemplo:  
```  
mivector = [ 56, 7, 34, 9, 23 ] # lista simple de 6 elementos.  
```   
**Matríz:**  Es una lista de elementos pero que se extiende en dos dimensiones, como una cuadrícula, por ejemplo: 

```  
mimatriz = [ [9, 24, 7], [13, -5, 42], [81, 43, 2] ]  # matriz de 3x3 elementos.  
```  
Esa matriz se vería algo así: 
```  
[
[ 9, 24, 7 ], 
[ 13, -5, 42],
[ 81, 43, 2]  
]
```  
Entonces las imágenes son lo mismo, una matriz de valores (de grises, rgb, Hexadecimal o el espacio de color que sea). Al ser una matriz de números, se pueden realizar muchas operaciones de transformación para modificar pixeles (escalar, rotar, etc).   

Esas es lo que hacen programas como photoshop, modificar pixeles aplicando transformaciones a los valores y a la matriz.

## C- Imagen: Operaciones. 

Al ser números, se pueden realizar operaciones de transformación a la imagen, es decir, a la matríz. Estas transformaciones son muy conocidas en diseño: desplazar, rotar, escalar (ampliar-reducir), etc. Todas operan en la matriz mediante operaciones matemáticas, que afortunadamente ya vienen definidas en algunas librerías de python. 


````python
# C- Imagen: Operaciones. Ejemplo Rotación. 

# Rotar matriz. 
fig, axflor = plt.subplots(figsize=(5, 5)) # tamaño
tr = transforms.Affine2D().rotate_deg( 19.9 ) # rotacion en grados
axflor.imshow(imgflor, transform=tr + axflor.transData)
#axflor.axis('off')
plt.show()
````

````python
# Recortar 
# para recortar tienen que ver cual es la forma de la imagen
# para estar dentro de los parametros de ancho y alto
# sabemos que imgflor tiene 250x250 pixeles. 

# en imgflor_recortada guardamos imgflor recortada
imgflor_recortada = imgflor[ 0:100, 0:100 ] # alto,ancho

fig, axrand = plt.subplots(figsize=(4, 4)) # tamaño
plt.imshow(imgflor_recortada)
plt.axis("off")
plt.show()
````

## D- Pixeles: Valores de Pixeles. 

Como vimos, una imagen es una matriz, y cada valor en dicha matriz es un número que también puede "leerse" como valor de color. Se pueden cambiar los valores globales de la manera en que se "leen" los valores de pixel, lo que afecta en cómo se mostrará una imagen. Para esto se puede usar la librería MatPlotLib (que debe cargarse al inicio) y seleccionar los espacios de color que interesan usar. 

También se puede seleccionar mostrar cada uno de los canales de la imagen. Por ejemplo si la imagen es RGB, se puede seleccionar mostrar el canal r, g, o el b. 


````python
# D- Pixeles: Ejemplo seleccionar mostrar uno de los cabanales RGB

# Se selecciona 1 canal, de los 3. [x, y, canal N ] 
imgflor_cR = imgflor[:, :, 0] # [ x, y, canal 0 ]
imgflor_cG = imgflor[:, :, 1] # [ x, y, canal 1 ]
imgflor_cB = imgflor[:, :, 2] # [ x, y, canal 2 ]

# Mostrar los valores de cada canal 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(imgflor_cR, cmap="gray" )
ax2.imshow(imgflor_cG, cmap="gray" )
ax3.imshow(imgflor_cB, cmap="gray" )

ax1.axis('off') # comentar para no mostrar los ejes 
ax2.axis('off') # comentar para no mostrar los ejes 
ax3.axis('off') # comentar para no mostrar los ejes 
````

````python
# E- Pixeles: Dibujar Pixeles. Ejemplo dibujar un punto, varios puntos de colores. 
# Markers: https://matplotlib.org/stable/api/markers_api.html

# Se dibujaran unos "markers" en la imagen
plt.plot(50, 50,  marker='o', color="red",    markersize=20)
plt.plot(50, 80,  marker='x', color="white",  markersize=20)
plt.plot(50, 110, marker='v', color="orange", markersize=20) 

# mostrar 
plt.imshow(imgflor)
plt.show()
````

## F- Pixeles: Visualizar distribución con Histograma. 

Los histogramas seguramente es algo que han visto en programas de edicion de imagen como PShop. 

Las imágenes tienen ancho y alto, lo que da una cantidad de pixeles de la imagen (alto x ancho) y cada uno de estos pixeles puede tener un valor en RGB. A veces es pertinente mostrar en una visualización para ver como se distribuyen los valores de pixeles, por ejemplo para ver la predominancia de ciertos colores en una imagen. Para esto se usa un gráfico llamado histograma, que cuenta cuantas veces se repite un valor.   


````python
# F- Visualizar distribución con Histograma.
# Por ejemplo el valor 150 aparece 10 veces en capa R, el 25 aparece 90 veces en capa G, etc...

# Se selecciona cada uno de los 3 canales. 
imgflor_cR = imgflor[:, :, 0] # [ todos los pixeles del r, todos los pixeles del g, cual canal ]
imgflor_cG = imgflor[:, :, 1]
imgflor_cB = imgflor[:, :, 2]

# Cada canal (RGB) se aplana de matriz 2d (anchoxalto), a un vector 1D. 
red_pxls = imgflor_cR.flatten()
gre_pxls = imgflor_cG.flatten()  
blu_pxls = imgflor_cB.flatten()  


# Para mostrar histograma de la imagen. 
# proporción del gráfico 
plt.figure(figsize=( 16, 10 ) )

# plt.hist(cual canal, bins=cuantas barritas, density=False, color=color de las barritas, alpha=transparencia de las barras)
plt.hist(red_pxls, bins=256, color='red',   alpha=0.4)
plt.hist(gre_pxls, bins=256, color='green', alpha=0.4)
plt.hist(blu_pxls, bins=256, color='blue',  alpha=0.3)

# Titulos
plt.title('Histograma Colores')
plt.ylabel('Cuantos por cada valor')
plt.xlabel('Intensidad de los valores de 0 a 255')

# Mostrar
plt.show()
````
