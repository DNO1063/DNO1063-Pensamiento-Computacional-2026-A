## QUÉ ES ESTADÍSTICA?   

**Estadística:** Describe un conjunto de datos obtenidos para tomar decisiones o realizar generalizaciones sobre características de las observacione. Usa un conjunto de métodos para obtener, analizar y presentar observaciones numéricas y/(o no numéricas.

**Dos ramas en Estadística:** La estadística se divide en dos grandes áreas, una describe lo dado, la otra busca predecir la ocurrencia de un fenómeno.

**A) Estadística descriptiva:** Resume datos de fenómenos usando métodos de recolección, descripción, visualización. Los datos pueden ser resumidos numérica o gráficamente. Usa medidas de descripción numérica como la media y la desviación estandar.   

**B) Estadística Inferencial:** Busca generalizar modelos, realizar inferencias y predicciones de fenómenos considerando la incertidumbre en las observaciones (probabilidad). Las inferencias pueden ser: estimaciones de características numéricas (estimación), pronósticos de futuras observaciones, descripciones de asociación (correlación), o relaciones entre variables (regresión), entre otras.

**Erorres más fecruentes en Estadística:** Datos de muestras no son representativos. Las conclusiones erróneas por datos insuficientes. Las representaciones gráficas pueden ser engañosas o distorsionadas (escalas). Ciertas categorías incorrectas. Respuestas o datos sesgadas.

## TIPOS DE DATOS
Según su naturaleza, los datos pueden ser:

**Cualitativos:** Valores de cualidades o atributos, clasificados en categorías. No son numéricos. Ej:  
* Sexo: f/m.
* Hábito de fumar: Fumador/No fumador
* Estado civil: soltero, casado, divorciado, ...

**Cuantitativos:** Valores que pueden cuantificarse o expresarse numéricamente. Ej:
*  Peso, Edad, Estatura: 1,60, 1,70, 180, ...
*  Presión, Humedad, Temperatura: 20, 30, 18, ...
*  Notas de un examen: 4.0, 5.5, 6.8, ...

**Datos discretos:** Datos que no pasan de una clase a la siguiente sin que haya una interrupción, representadas como números enteros. Ej:
* Estudiantes de PhD en una universidad.
* 2 gatos
* 2 líneas de metro  

**Datos continuos:** Datos que pueden pasar de una clase otra siguiente sin interrupción y que pueden expresarse mediante números enteros o fraccionarios. Puede tomar cualquier valor en una escala continua. Ej:
* Cantidad de líquido: 800, 800.5, ...
* Peso: 60.25, 70.2, ...
       
**Univariadas o unidimensional:** Recogen información sobre una característica. Ej:
* edad de alumnos.
* autos vendidos.

**Bivariado o bidimensional:** Recogen información sobre dos características de la población. Ej:
* edad y estatura de alumnos.
* autos vendidos y su marca.
    
**Multivariados o pluridimensional:** recogen información sobre tres ó más características. Ej:
* edad, estatura, y peso de alumnos.
* autos vendidos, su marca, y año.

## B- Medidas de tendencia central o posición.

Son valores que generalmente se ubican en la parte central de un conjunto de datos. Indican como los datos se condensan en un solo valor central, alrededor del cual los demás datos se distribuyen. Son la Media aritmética, Mediana, y la Moda.   

**Media Aritmética:** Es la suma de todas las observaciones dividida entre el total de observaciones. es una de las formas más usadas para resumir la información de una distribución, pero es a veces engañosa, ¿Por qué?

Ventajas: Es un concepto familiar para la mayoría de las personas. Permite la comparación de medias de varios conjuntos de datos.

Desventajas: Se ve afectada por valores extremos, perdiendo representatividad. No se puede calcular para datos cualitativos. "Si mi vecino tiene dos coches y yo ninguno, los dos tenemos uno”.

1, 3, 6, 2, 4, 2, 5  
Total= 23  
23/7 = 3.28  


**Mediana:** Valor que ocupa la posición central de un conjunto de datos, una vez ordenados en forma ascendente o descendente. Divide al conjunto de datos en dos partes iguales, par o impar.

Ventajas: Fácil de calcular. No se ve influida por valores extremos, solo influyen valores centrales. Fácil de entender.

Desventajas: No utiliza en su “cálculo” toda la información disponible. No pondera cada valor por el número de veces que se ha repetido. Hay que ordenar los datos antes de determinarla.

1, 3, 6, 2, 4, 2, 5  
1, 2, 2 (3) 4, 5, 6 - ordenados  


**Moda:** Observación que tiene la mayor frecuencia en un conjunto de datos. Un conjunto de datos puede ser unimodal, bimodal o multimodal.  

Ventajas: Fácil de interpretar. No requiere cálculos solo contar las que más se repiten. Puede usarse para datos cuantitativos y cualitativos. No se ve influida por valores extremos.

Desventajas: Tiene significado en el caso de una gran cantidad de datos. No utiliza toda la información disponible. No siempre existe, si los datos no se repiten. El azar puede hacer que una observación no representativa sea el valor más frecuente. Difícil de interpretar si los datos tienen 3 o más modas.

1, 3, 6, 2, 4, 2, 5
2 (se repite 2 veces)

## C- Medidas de dispersión, variación o variabilidad.  

* Son valores numéricos que indican o describen la forma en que las observaciones están dispersas respecto al valor central.
* Son importantes debido a que dos muestras de observaciones con el mismo valor central pueden tener una variabilidad muy distinta.
* Son: Rango, Varianza (no la veremos). Desviación Estándar.

#### Rango:
* Es la diferencia entre el dato mayor y el menor.
* Permite conocer la máxima dispersión.
* Se emplea poco debido a que depende únicamente de dos valores.
* Desventajas:
    * No es una medida de Dispersión con respecto al centro de la distribución.
    * Solo emplea dos valores en su cálculo.

1, 3, 6, 2, 4, 2, 5  
mayor = 6  
menor = 1  
rango (mayor - menor) = 5  



#### Desviación Estandar:
* Es la raíz cuadrada de la varianza.
* Permite ver que tan lejos está un valor del centro.
Ventajas:
* Expresada en las mismas unidades que la variable en estudio.
* Utiliza todas las observaciones en su cálculo.
* Fácil de interpretar.
Desventajas:
* No tiene.

1, 3, 6, 2, 4, 2, 5  
media 3.28
1 esta a 2.28 std  
6 esta a 2.71 std  



```python
## Maximo
max = df['juegos_ganados'].max(numeric_only=True)
print("Valor Máximo: ", max )

## Minimo
min = df['juegos_ganados'].min(numeric_only=True)
print("Valor Mínimo: ", min )
```

```python
## Rango de variable juegos ganados.
rango = df['juegos_ganados'].max() - df['juegos_ganados'].min()
print( "Rango:", rango )
```

```python
## Desviacion Estandar

desv_est = df['juegos_ganados'].std()
print( "desviacion_estandar:", desv_est )
```

## E- Graficando algunas variables: Scatterplot y Torta.   
    
* En este scaterplot graficaremos una variable o serie.
* Usted puede probar agregando títulos, cambiando colores, tamaños, guardando el gráfico, etc.
* Marcadores de puntos - https://matplotlib.org/stable/api/markers_api.html


```python
## Scatterplot de una serie.

## Variables.
x1 = df['team'] # en x1 la variable dependiente
y1 = df['goles_favor'] # en y1 guardaremos la variable dependiente

#plt.figure(figsize=(14,10))
plt.scatter(x=x1, y=y1, color='red') # gráfico de puntos.
```

```python
## Scatterplot de dos series
## En este scatterplot graficaremos dos variables o series.

## Datos - guardar datos en dos variables
d1 = (df['team'], df['goles_favor'])
d2 = (df['team'], df['goles_contra'])

sdatos = (d1, d2) # guardamos esos grupos en la variable datos.
snombres = ('Goles a favor', 'Goles en contra') # guardamos esos grupos en la variable datos.
scolores = ('red', 'dodgerblue') #


### Crear grafico - - - - - - - - - - - -
fig=plt.figure(figsize=(18, 10) ) # tamaño del gráfico.
fig.patch.set_facecolor('green') # color del margen.
fig.patch.set_alpha(0.9) # transparencia del margen.

ax = fig.add_subplot(1, 1, 1) # sub gráficos en el mismo espacio.
ax.patch.set_color('fuchsia') # transparencia del gráfico.
ax.patch.set_alpha(0.9) # transparencia del gráfico.


## Titulos y colores - - - - - - - - - - - -
plt.title("Relación Goles Favor - Goles Contra'") # título.
plt.xticks(rotation=0)
plt.xlabel("Equipos") # texto eje X.
plt.ylabel("Goles") # texto eje Y.
plt.grid( color='blue', alpha=0.9) # grilla.

## Ciclo recorre sdatos, scolores, snombres y se guardan en dat, col, gru.
for dat, col, gru in zip(sdatos, scolores, snombres):
    x, y = dat
    ax.scatter(x, y, alpha=0.6, c=col, edgecolors='none', s=180, label=gru, marker='D') # puntos.
    plt.legend(snombres) # leyendas


## Mostrar y guardar gráfico - - - - - - - - - - - -
# guarda el gráfico en la ruta seleccionada.
#plt.savefig('guardar/c_cuarteto_scatter_series.jpg')
# esta linea muestra el gráfico.
plt.show()
```
