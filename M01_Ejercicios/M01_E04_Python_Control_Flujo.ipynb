## A. Variables y Tipos de datos: Guardar valores.    

Se puede pensar en las variables como una cajita para guardar valores. Las variables crean un espacio en la memoria del computador para almacenar valores. Las variables se pueden pensar en 3 pasos: declarar, asignar, usar.

Se pueden realizar operaciones aritméticas (+ - * /), entre otras, también se puede incrementar (aumentar) y decrementar (disminuir).

Los tipos de datos más comunes son: `int` (entero), `float` (decimal), `str` (texto) y `bool` (True/False). Para combinar texto con variables se puede usar un f-string.

```python
# Ejemplo A - Variables y Tipos de datos

# operaciones
a = 15
b = 20
suma  = a + b
resta = b - a
multi = a * b
divis = b / a

print( suma )
print( resta )
print( multi )
print( divis )


# incrementar y decrementar
a = 15
a = a + 1  # incrementar
a = a - 1  # decrementar


# tipos de datos
mi_entero = 10
mi_float  = 20.5
mi_str    = "hola"
mi_bool   = True

print( mi_entero + mi_float )  # sumar dos valores
print( mi_str )


# f-string
nombre = "Ana"
print( f"Hola, {nombre}" )
```

## B. If-else: Tomar decisiones.     

El computador puede tomar decisiones según ciertas condiciones, comparando valores. Para esto se usan ciertas estructuras como if, else, elif. Estas se pueden pensar como una pregunta con una respuesta de "sí" o "no", True o False.

Para comparar variables se usan operadores relacionales:

* `>` Mayor
* `<` Menor
* `>=` Mayor Igual
* `<=` Menor Igual
* `!=` Diferente
* `==` Igual, ojo: un = es asignar, dos == es comparar.

```python
# Ejemplo B - If-else

# 2a- if simple
luz = 1  # probar cambiando el valor
if luz == True:
    print( "2a: la luz está encendida" )


# 2b- if / else
a = 10
b = 20
if a < b:
    print( "2b: a es menor" )
else:
    print( "2b: b es menor" )


# 2c- if / elif / else
hambre = True
dinero = True

if hambre and dinero:
    print( "2c: Juan se come un sandwich" )
elif hambre and not dinero:
    print( "2c: Juan pide plata a su mamá" )
else:
    print( "2c: Juan se queda con hambre" )
```

## C. Ciclos For: Repetir tareas.  

Los ciclos, loops o iteraciones, son estructuras que, mientras se cumpla una condición, repiten un proceso n veces. Por ejemplo, se establece un límite o rango, y mientras no se llega a ese valor, se repite la acción, y al llegar a ese valor límite, se detiene. La variable i (u otra letra escogida) va cambiando a cada iteración: 0, 1, 2, 3... Ese valor se puede usar para recorrer listas.

```python
# Ejemplo C - Ciclos For

sep = "- - - - - - - - - - - - - - - - - - - - - - - - "


# 3a- rango simple
for x in range( 4 ):
    print( "x vale:", x )
print( sep )


# 3b- sumando valores
for x in range( 4 ):
    print( "x vale:", x + 5 )
print( sep )


# 3c- iterar en una lista
meses = ["ene", "feb", "mar", "abr", "may", "jun", "jul", "ago", "sep", "oct", "nov", "dic"]
for i in range( len( meses ) ):
    print( meses[i] )
```

## D. Listas y Arrays: Guardar colecciones de variables.   

Los arrays, o listas, sirven para guardar muchos valores. Es una colección de cajas para guardar datos. Al igual que las variables, se declaran, se le asignan valores, y se usan. Los valores están separados por coma. Se accede a los datos por medio de su index o posición. Considerar que en Python la primera posición es 0. Se puede conocer su largo con `len()`.

Las listas pueden contener tipos de datos diferentes, e incluso listas dentro de listas. Para acceder a un elemento de una lista anidada se usan dos índices, por ejemplo `lista[0][1]`.

```python
# Ejemplo D - Listas
# Se pueden guardar diferentes tipos de datos en una lista, se accede por su index

miArray = ["buen dia", 17.5, "buenas noches"]
print( miArray[0] )
print( miArray[1] )
print( miArray[2] )


# lista con más elementos
nombres = ["Juan", "Pedro", "Maria", "Clara", "Rogelio"]
print( len( nombres ) )  # largo de la lista


# lista dentro de lista
l1 = [ [1, 2], [3, 4] ]
print( l1[0] )      # muestra la primera lista
print( l1[0][1] )   # muestra el segundo elemento de la primera lista
```
