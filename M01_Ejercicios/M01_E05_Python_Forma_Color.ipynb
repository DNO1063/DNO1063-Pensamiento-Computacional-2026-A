## C. Color Hexadecimal.

En programación, el color se puede representar en sistema hexadecimal, que es de base 16. Usa los dígitos del 0 al 9 y las letras A, B, C, D, E, F (donde A=10, B=11... F=15). Con combinaciones de 6 dígitos se pueden representar más de 16 millones de colores.

Los colores hexadecimales llevan # al inicio y van en pares: #RRGGBB, donde el primer par es rojo, el segundo verde y el tercero azul. Por ejemplo, #FF8C00 es naranja.

Se pueden ver colores en: https://www.w3schools.com/colors/colors_hexadecimal.asp

```python
# Ejemplo C - Colores hexadecimales guardados en variables.

col_amar = "#FFFF00"
col_azul = "#0000FF"
col_blan = "#FFFFFF"
col_negr = "#000000"
col_rojo = "#FF0000"
col_verd = "#008000"
col_nara = "#FF8C00"

print( col_amar )
```

## D. Formas y Texto.

Con Matplotlib se pueden dibujar formas básicas: rectángulo, círculo, línea, y también agregar texto. Cada forma se crea con una función de `matplotlib.patches` y se agrega al gráfico con `ax.add_patch()`. Al final se usa `plt.show()` para mostrar el resultado.

Las unidades de posición y tamaño las define `plt.xlim()` y `plt.ylim()`.

```python
# Ejemplo D - Formas y colores con Matplotlib

fig = plt.figure( figsize=(9, 6) )
ax  = fig.add_subplot(111)

# Rectangle( (x, y), ancho, alto, color=hexadecimal )
rect1 = matplotlib.patches.Rectangle( (10, 10), 80, 20, color=col_amar )
rect2 = matplotlib.patches.Rectangle( (0, 25),  100, 25, color=col_rojo )

# Circle( (x, y), radio, color=hexadecimal )
circ1 = matplotlib.patches.Circle( (50, 40), radius=10, color=col_verd )

ax.add_patch( rect1 )
ax.add_patch( rect2 )
ax.add_patch( circ1 )

# text( x, y, "texto", size=px, rotation=grados )
plt.text( 35, 16, "gato", size=20, rotation=0 )

plt.xlim( [0, 100] )
plt.ylim( [0, 75]  )
plt.show()
```
