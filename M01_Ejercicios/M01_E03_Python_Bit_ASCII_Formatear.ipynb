## A - Números binarios

El computador trabaja con bits, la unidad de datos más básica. Un bit vale 0 o 1. Ocho bits forman un byte.

Los bits tienen un valor posicional que va aumentando por dos de derecha a izquierda: 128, 64, 32, 16, 8, 4, 2, 1. Por ejemplo, `00100101` en binario equivale a 37 en decimal.

En Python, un número binario se escribe con el prefijo `0b`: `0b00100101`.

```python
# Ejemplo A - número binario
print( 0b00100101 )  # muestra 37
print( 0b01001101 )  # ¿cuánto es este?
```

## B - Codificación ASCII

Los caracteres (letras, números, símbolos) están representados internamente como números. ASCII es el sistema de codificación que asigna un número a cada caracter.

En Python:

- `ord(caracter)` convierte un caracter a su número ASCII
- `chr(numero)` convierte un número ASCII a su caracter

```python
# Ejemplo B - ord() y chr()
print( ord("a") )  # 97
print( chr(97) )   # a
print( ord("A") )  # 65
```

## C - Formatear texto

Al poner un texto entre triple comillas (`"""..."""`), se conserva el formato y los saltos de línea tal como se escriben.

También existen caracteres especiales que no se ven en pantalla:

- `\n` produce un salto de línea
- `\t` produce una tabulación (sangría)

```python
# Ejemplo C1 - texto multilínea con triple comillas
lineas = """Mi nombre
    es Ana
        estudio
    diseño"""

print( lineas )
```

```python
# Ejemplo C2 - salto de línea con \n
frase = "No \nlas \ndamas, \namor"
print( frase )
```
