## A - Print y variables

La función `print()` muestra valores en la consola, el espacio que aparece bajo cada celda. Sirve para comprobar si los valores son los que esperamos.

Se puede pasar directamente un valor, una variable, o varios separados por comas.

```python
# Ejemplo A - print con variables
a = "*   * *****  **** *****"
b = "*   * *     *     *   *"
c = "*   * ***** *  ** *****"
d = " * *  *     *   * *   *"
e = "  *   *****  ***  *   *"

print( a )
print( b )
print( c )
print( d )
print( e )
```

## B - Sumatoria

Sumar todos los valores de una lista se llama sumatoria. Se puede hacer de dos formas: sumando variable por variable, o recorriendo una lista con un ciclo `for` y acumulando el resultado.

La variable acumuladora parte en 0 y se va sumando en cada paso del ciclo.

```python
# Ejemplo B1 - sumar con variables individuales
r01 = 1
r02 = 0
r03 = 2
r04 = 3
r05 = 5

tot = r01 + r02 + r03 + r04 + r05
print( "La suma es:", tot )
```

```python
# Ejemplo B2 - sumar desde lista con ciclo for
rut   = [1, 0, 2, 3, 5, 0, 2, 8, 9]
tot_b = 0

for i in range( len(rut) ):
    tot_b = tot_b + rut[i]

print( "La suma es:", tot_b )
```

## C - Concatenar

Concatenar es unir textos y números para armar una frase. No se puede sumar directamente un número con un string; primero hay que convertir el número a string con `str()`.

También se puede usar `print()` con comas para mostrar varios valores juntos, o un f-string: `f"texto {variable}"`.

```python
# Ejemplo C - concatenar texto y número
nombre = "Juan"
num    = 45

frase  = "Me llamo " + nombre + ", mi número es " + str(num)
print( frase )
```
