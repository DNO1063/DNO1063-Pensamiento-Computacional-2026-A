## A - Variables y tipos de datos

Una variable es un nombre con el que guardamos un valor para usarlo después.

Los tipos de datos más comunes son:

- `int`: números enteros como 42, -3
- `float`: números decimales como 2.87, 3.14
- `str`: texto entre comillas como "hola", "Python"
- `bool`: valores lógicos, True o False

Se pueden hacer operaciones aritméticas con números: `+`, `-`, `*`, `/`.
Para saber el tipo de un valor se usa `type()`.

```python
# Ejemplo A - variables y operaciones
nombre = "Ana"
edad   = 21
altura = 1.65

print( nombre )
print( edad + 5 )   # operación aritmética
print( type(edad) ) # int
```

## B - Decisiones (if/else)

El computador puede tomar decisiones según si una condición es verdadera o falsa.

La estructura básica es `if condición:` seguido del bloque a ejecutar si es verdadero, y `else:` para el caso contrario.

Operadores de comparación: `==` igual, `!=` distinto, `>` mayor, `<` menor.

```python
# Ejemplo B - decisión con if/else
a = 3

if a == 5:
    print( "a vale 5" )
else:
    print( "a NO vale 5" )
```

## C - Ciclos (for)

Un ciclo `for` repite instrucciones una cantidad determinada de veces.

`range(n)` genera los números desde 0 hasta n-1. La variable del ciclo toma cada uno de esos valores en cada repetición.

```python
# Ejemplo C - ciclo for con range
for i in range( 6 ):
    print( "i vale:", i )
```

## D - Listas

Una lista guarda varios valores bajo un mismo nombre. Cada elemento tiene una posición llamada index, que parte en 0.

Para acceder a un elemento: `lista[index]`.

```python
# Ejemplo D - lista y acceso por index
nombres = ["Ana", "Beatriz", "Claudia", "Diana", "Elvira"]

print( nombres[0] )  # primer elemento
print( nombres[4] )  # quinto elemento
```

## E - Recorrer listas

Se puede recorrer una lista con un ciclo `for`. `len(lista)` entrega la cantidad de elementos de la lista.

Hay dos formas: recorrer con índices usando `range(len(lista))`, o recorrer directamente los elementos.

```python
# Ejemplo E - recorrer lista con índices
nombres = ["Ana", "Beatriz", "Claudia", "Diana", "Elvira"]

for i in range( len(nombres) ):
    print( i, nombres[i] )
```
